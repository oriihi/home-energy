```python
import pandas as pd
import numpy as np
import holoviews as hv
import hvplot
import hvplot.pandas
```

```python
df = pd.read_csv('./data/20210310-water-pump.csv',
                 index_col='time',
                 parse_dates=True)
```

```python
df.drop(columns='#',inplace=True)
df.drop(columns='End Of File (LGR S/N: 1191118)',inplace=True)
df.head()
```

```python
df.hvplot.line(grid=True)
```

```python
# Convert to Whr's used
samples_per_hour = 60*60
water_pump_voltage = 220

df['water_pump_wh'] = (df['water_pump'] / samples_per_hour) * water_pump_voltage
df['water_pump_wh_accum'] = df['water_pump_wh'].cumsum()
```

```python
df.water_pump_wh_accum.hvplot.line(grid=True) 
```
